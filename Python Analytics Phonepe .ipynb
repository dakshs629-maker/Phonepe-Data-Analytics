```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# ==========================================
# PART 1: THE HYPOTHESIS TEST (The Math)
# ==========================================
# This is the "Bootstrapping" technique from your PDF.
# It simulates the experiment 5,000 times to find the "True Mean."

# 1. Get the raw data
loans = df[df['Service'] == 'Loans']['Amount']
insurance = df[df['Service'] == 'Insurance']['Amount']

# 2. Define the Bootstrap Function (The Logic)
def bootstrap_means(data, n_reps=5000):
    means = np.empty(n_reps)
    for i in range(n_reps):
        # Resample with replacement (The "Test")
        sample = np.random.choice(data, size=len(data))
        means[i] = np.mean(sample)
    return means

# 3. Run the Test (Generate the Data)
# These variables (bs_loans, bs_insurance) hold the results of your hypothesis test
bs_loans = bootstrap_means(loans, n_reps=5000)
bs_insurance = bootstrap_means(insurance, n_reps=5000)

# ==========================================
# PART 2: THE VISUALIZATION (The Graph)
# ==========================================
# This takes the math from Part 1 and draws the curves.

plt.figure(figsize=(12, 6))

# Plot 1: Insurance Distribution
sns.kdeplot(bs_insurance, color='#E74C3C', fill=True, alpha=0.6, linewidth=2, label='Insurance (Lower Ticket Size)')

# Plot 2: Loans Distribution
sns.kdeplot(bs_loans, color='#2ECC71', fill=True, alpha=0.6, linewidth=2, label='Loans (Higher Ticket Size)')

# Formatting
plt.title('Hypothesis Test: The "Value Gap" between Loans & Insurance', fontsize=16, fontweight='bold')
plt.xlabel('Average Transaction Value (Rupees)', fontsize=12)
plt.ylabel('Probability Density', fontsize=12)

# Add the "Truth" Arrow
plt.annotate('Zero Overlap = Statistical Certainty', 
             xy=(25000, 0.00005), 
             xytext=(25000, 0.00015),
             arrowprops=dict(facecolor='black', shrink=0.05),
             fontsize=10, ha='center')

plt.legend(fontsize=12)
plt.grid(True, alpha=0.2, linestyle='--')
plt.show()
```

```python
# Level 5: Strategic Sector Scorecard (Bloomberg Style)
# -----------------------------------------------------
# Why this matters: Combines Growth, Profit, and Reliability into one view.

# 1. Aggregate key metrics
scorecard = df.groupby('Service').agg({
    'Amount': ['mean', 'sum'],              # Profitability & Total Revenue
    'Transaction_ID': 'count',              # Volume
    'Payment_Status': lambda x: (x == 'Successful').mean()  # Success Rate
})

# 2. Rename columns for clarity
scorecard.columns = ['Avg_Ticket_Size', 'Total_Revenue', 'Volume', 'Success_Rate']

# 3. Normalize (0 to 1 scale) so we can compare "Apples to Oranges"
# (e.g., Comparing $2 Billion Revenue vs. 96% Success Rate)
scorecard_normalized = (scorecard - scorecard.min()) / (scorecard.max() - scorecard.min())

# 4. Plot the Strategic Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(scorecard_normalized, annot=True, cmap='RdYlGn', fmt='.2f', linewidths=1)

plt.title('Strategic Sector Scorecard: Relative Performance Matrix', fontsize=14)
plt.ylabel('Business Unit', fontsize=12)
plt.show()
```
